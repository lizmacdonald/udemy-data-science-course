```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
plt.plot(x, y)
#plt.show() ##use in commandline
```

```python
plt.plot(x, y,)
plt.xlabel('x Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
## functional method of plot creation
#plt.subplot(nuber of rows, number of columns, plot number youre referring to)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
## object oriented method
## axis takes in a left, bottom, width, and hieght argument
## axis arguments range from 0-1

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

##plot on those axies
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGE')

axes2.plot(y, x)
axes2.set_title('SMALL')
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)
#axes.plot(x, y)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

#for current_ax in axes:
 #   current_ax.plot(x, y)
    
axes[0].plot(x, y)
axes[1].plot(y, x)

axes[0].set_title("first")
axes[1].set_title("second")

plt.tight_layout() ##takes care of overlap of plots
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = "x squared")
ax.plot(x, x**3, label = "X cubed")

## add in legend can specify location based on a code number
## or can input a tuple to give the x,y coord of where you want it to be

ax.legend(loc = (0.1, 0.1))
```

### setting colors

## for colors, line style and marker options, go to notebook sheet

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = "purple")  ##can also input RGB hex code to generate more specific colors
```

```python

##line width
##linewideth default is 1
# alpha is the transpamcy of the line
## can use lw instead of linewidth
## you can use ls instead of linestyle
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = "purple", lw = 3, ls = "--")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = "purple", lw = 1, ls = "-", marker = "o", markersize = 30,
       markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'green')
```

```python
## only want to show between 0 and 1 on the x axis
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = "purple", lw = 1, ls = "-")
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
